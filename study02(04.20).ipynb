```python
import os
import csv
import numpy as np
import pandas as pd
from konlpy.tag import Hannanum
import re
import pytagcloud
import random
import webbrowser
from konlpy.tag import Twitter
from collections import Counter
import matplotlib.font_manager as fm
import numpy as np
import random
from PIL import Image
hn = Hannanum()

os.getcwd()

os.chdir(r'C:\Users\jmlee\Desktop\4학년 1학기\텍스트마이닝을 이용한 소비자분석\0420 실습')

def flatten(l):
    flat_list = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flat_list.append(e)
        else:
            flat_list.append(elem)
    return flat_list

f = open("Boki.txt", 'r', encoding='UTF8')
lines = f.readlines()
f.close()

#불용어 제거
# lines = re.sub('[0-9]+', '', str(lines))
# lines = re.sub('[A-Za-z]+', '', str(lines))
# lines = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', str(lines))
hangul = re.compile('[^ ㄱ-ㅣ가-힣]+')  #한국어만 출력
result = hangul.sub('', str(lines))
result
```
