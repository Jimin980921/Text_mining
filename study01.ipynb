```python
#1번.
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
os.getcwd()

os.chdir('C:\\Users\\jmlee\\Desktop\\4학년 1학기\\텍스트마이닝을 이용한 소비자분석\\0406 실습')
f = open('202002_ages.csv')   #csv파일 가져오기
data=csv.reader(f)  #csv를 읽어서 data에 대입
result1 = []
result2 = []

name1= input('동 이름을 입력하세요: ')
name2= input('동 이름을 입력하세요: ')

for row in data :
    if name1 in row[0] :   #지역
        for i in row[3:] :  #인구수 가져오기
            result1.append(int(i.replace(',', '')))
            
    elif name2 in row[0]:
        for i in row[3:]:
            result2.append(int(i.replace(',', '')))
    
    
# for row in data :
#     if name2 in row[0] :   #지역
#         for i in row[3:] :  #인구수 가져오기
#             result2.append(int(i.replace(',', '')))
#             print(result2)


plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')

plt.title(name1+','+name2+'의 인구비교') 
name1=str(name1)
name2=str(name2)
plt.plot(result1,label=name1)
plt.plot(result2,label=name2)
plt.legend(loc='upper right') #라벨표시

plt.show()
```
