```python
#2번.

import os
import csv
import numpy as np
import pandas as pd
from konlpy.tag import Hannanum
hn = Hannanum()

os.getcwd()

os.chdir('C:\\Users\\jmlee\\Desktop\\4학년 1학기\\텍스트마이닝을 이용한 소비자분석\\0406 실습')

#리스트 flat 함수
def flatten(l):
    flat_list = []
    for elem in l:
        if type(elem) == list:
            for e in elem:
                flat_list.append(e)
        else:                                 
            flat_list.append(elem)
    return flat_list


#문서 로딩
f = open("2020년31절_대통령연설문.txt", 'r', encoding='UTF8')
lines = f.readlines()
f.close()

#명사추출
temp = []
for i in range(len(lines)):
    temp.append(hn.nouns(lines[i]))


#빈도계산
word_list = flatten(temp)
word_list = pd.Series([x for x in word_list if len(x)>1])
word_list.value_counts().head(15)
```
